```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import uuid   # Unique identifier
import os
import time
```

```python
images_path = os.path.join('data', 'images') #/data/images
labels = ['sad', 'happy', 'angry']
number_images = 202
```

```python
for label in labels:
    print(f"Collecting images for {label}")
    for image_number in range(number_images):
        print(f"Collecting images for {label}, image number {image_number}")
        image_name = os.path.join(images_path, label+'.'+str(uuid.uuid1())+'.jpg')
        print(image_name) 
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp5/weights/last.pt', force_reload=True)
```

```python
image_sad = os.path.join('data', 'images', 'janko-ferlic-oWDRVgk04EA-unsplash.jpg')
result_sad = model(image_sad)
result_sad.print()

image_happy = os.path.join('data', 'images', 'eye-for-ebony-OWi1sIWiCAI-unsplash.jpg')
result_happy = model(image_happy)
result_happy.print()

image_angry = os.path.join('data', 'images', 'AdobeStock_247409832_Preview.jpeg')
result_angry = model(image_angry)
result_angry.print()
```
